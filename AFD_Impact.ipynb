```python
def match(actor: str, dictionary: dict):
    actor = actor.lower()
    nrOfMatches = 0
    rating = 0
    for key in dictionary.keys():
        if actor in stringToList(dictionary[key]):
            rating += rating_dict[key]
            nrOfMatches += 1
    if nrOfMatches == 0:
        return 0
    else:
        return rating/nrOfMatches


def matchData(actors: list, dictionary: dict):
    score = 0
    for actor in actors:
        score += match(actor, dictionary)
    return score / len(actors)


def stringToList(string: str):
    string = string.lower()
    if string.find(",") == -1:
        return [string]
    else:
        return string.replace(", ", ",").split(",")
```

```python
def removeEndLine(listOfWords: list):
    for i in range(len(listOfWords) - 1):
        listOfWords[i] = " " + listOfWords[i].replace("\n", " ").replace("\r", " ")
    return listOfWords


def filterSigns(string: str):
    for i in ".?,!;:+=%*":
        string = string.replace(i, "")
    return string


def filterWords(inputString: str):
    with open("input.txt") as file:
        listOfWords = file.readlines()
        listOfWords = removeEndLine(listOfWords)
    for word in listOfWords:
        inputString = inputString.replace(word, " ")
    return inputString


def matchDescription(v: str, setOfWords: set):
    score = 0
    v = " " + filterSigns(v).lower() + " "
    for word in setOfWords:
        string = " " + word + " "
        if v.find(string) != -1:
            score += 1
    return score


def rateDescription(inputDict: dict, description: str):
    scores = dict()
    description = filterSigns(description).lower()
    list = filterWords(description).split(" ")
    setOfWords = set(list)
    for i in inputDict.keys():
        scores[i] = pow(matchDescription(inputDict[i], setOfWords),10)
    return scores
```

```python
def rating_descriptions(description_rating: dict):
    sum_description = sum(description_rating.values())
    for key in description_rating: 
        description_rating[key] = description_rating[key]/sum_description

    total_sum = 0    
    for key in description_rating:
        total_sum += rating_dict[key]*description_rating[key]
    return total_sum
```

```python
def movie_rating(director, writer, companyname, actor, language, country, genre, description):
    
    director_ = matchData(stringToList(director), director_dict)
    writer_ = matchData(stringToList(writer), writer_dict)
    actor_ = matchData(stringToList(actor), actors_dict)
    company_name_ = matchData(stringToList(companyname), company_dict)
    language_ = matchData(stringToList(language), language_dict)
    country_ = matchData(stringToList(country), country_dict)
    genre_ = matchData(stringToList(genre), genre_dict)
    description_ = rating_descriptions(rateDescription(description_dict, description))
    return (5*director_+ 5*writer_+ 7*company_name_ + 12.5*actor_ + 5*language_ + 2*country_ +12.5*genre_ + 50*description_)*0.01
```
